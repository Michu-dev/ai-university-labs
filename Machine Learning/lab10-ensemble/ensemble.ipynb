```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, VotingClassifier, StackingClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import GridSearchCV, StratifiedKFold, train_test_split
from imblearn.metrics import geometric_mean_score
from sklearn.metrics import accuracy_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.preprocessing import StandardScaler, Normalizer
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
```

```python
pd.set_option('display.max_columns', 0)
pd.set_option('display.max_rows', 0)
```

```python
classes = dict(
    ada_boost_gnb=lambda: AdaBoostClassifier(GaussianNB()),
    ada_boots_tree=lambda: AdaBoostClassifier(DecisionTreeClassifier()),
    voting_tree_svc=lambda: VotingClassifier([('tree', DecisionTreeClassifier()), ('svc', SVC())], n_jobs=4),
    stacking_tree_svc=lambda: StackingClassifier([('tree', DecisionTreeClassifier()), ('svc', SVC())], n_jobs=4),
    voting_svc_gnb=lambda: VotingClassifier([('svc', SVC()), ('gnb', GaussianNB())], n_jobs=4),
    stacking_svc_gnb=lambda: StackingClassifier([('svc', SVC()), ('gnb', GaussianNB())], n_jobs=4),
    voting_tree_gnb_svc=lambda: VotingClassifier([('tree', DecisionTreeClassifier()), ('gnb', GaussianNB()), ('svc', SVC())], n_jobs=4),
    stacking_tree_gnb_svc=lambda: StackingClassifier([('tree', DecisionTreeClassifier()), ('gnb', GaussianNB()), ('svc', SVC())], n_jobs=4),
    voting_tree_svc_gnb=lambda: VotingClassifier([('tree', DecisionTreeClassifier()), ('svc', SVC()), ('gnb', GaussianNB())], n_jobs=4),
    stacking_tree_svc_gnb=lambda: StackingClassifier([('tree', DecisionTreeClassifier()), ('svc', SVC()), ('gnb', GaussianNB())], n_jobs=4),
    random_forest=lambda: RandomForestClassifier(n_jobs=4)
)

fold = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
res = []
for name, cls in classes.items():
    for tr_idx, te_idx in fold.split(X, y):
        X_tr, X_te = X.loc[tr_idx], X.loc[te_idx]
        y_tr, y_te = y.loc[tr_idx], y.loc[te_idx]
        model = cls()
        model.fit(X_tr, y_tr)
        y_pr = model.predict(X_te)
        res += [dict(name=name, gmean=geometric_mean_score(y_te, y_pr))]
res = pd.DataFrame(res)
res
```

```python
vals = res.groupby('name').agg({'gmean': ['mean', 'std']})
vals
```

```python
plt.figure(figsize=(20, 12))
plt.bar(vals.index, vals.gmean['mean'].values, yerr=vals.gmean['std'].values, capsize=10)

plt.xlabel('Classifier name')
plt.ylabel('Geometric mean score')
plt.title('Geometric mean score for imbalanced data using ensemble classifiers')
plt.xticks(rotation=90)
plt.savefig('wykres_gmean_zadanie_4.png')
```

```python
classes = dict(
    ada_boost_gnb=lambda: AdaBoostClassifier(GaussianNB()),
    ada_boots_tree=lambda: AdaBoostClassifier(DecisionTreeClassifier()),
    voting_tree_svc=lambda: VotingClassifier([('tree', DecisionTreeClassifier()), ('svc', SVC())], n_jobs=4),
    stacking_tree_svc=lambda: StackingClassifier([('tree', DecisionTreeClassifier()), ('svc', SVC())], n_jobs=4),
    voting_svc_gnb=lambda: VotingClassifier([('svc', SVC()), ('gnb', GaussianNB())], n_jobs=4),
    stacking_svc_gnb=lambda: StackingClassifier([('svc', SVC()), ('gnb', GaussianNB())], n_jobs=4),
    voting_tree_gnb_svc=lambda: VotingClassifier([('tree', DecisionTreeClassifier()), ('gnb', GaussianNB()), ('svc', SVC())], n_jobs=4),
    stacking_tree_gnb_svc=lambda: StackingClassifier([('tree', DecisionTreeClassifier()), ('gnb', GaussianNB()), ('svc', SVC())], n_jobs=4),
    voting_tree_svc_gnb=lambda: VotingClassifier([('tree', DecisionTreeClassifier()), ('svc', SVC()), ('gnb', GaussianNB())], n_jobs=4),
    stacking_tree_svc_gnb=lambda: StackingClassifier([('tree', DecisionTreeClassifier()), ('svc', SVC()), ('gnb', GaussianNB())], n_jobs=4),
    random_forest=lambda: RandomForestClassifier(n_jobs=4)
)

fold = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
res = []
for name, cls in classes.items():
    for tr_idx, te_idx in fold.split(X, y):
        for norm in [False, True]:
            X_tr, X_te = X.loc[tr_idx], X.loc[te_idx]
            y_tr, y_te = y.loc[tr_idx], y.loc[te_idx]
            if norm:
                scaler = StandardScaler()
                X_tr = scaler.fit_transform(X_tr)
                X_te = scaler.transform(X_te)
            model = cls()
            model.fit(X_tr, y_tr)
            y_pr = model.predict(X_te)
            res += [dict(name=name, gmean=geometric_mean_score(y_te, y_pr), norm=norm)]
res = pd.DataFrame(res)
res
```

```python
import seaborn as snb
plt.figure(figsize=(20, 12))
plt.style.use('ggplot')
sns.barplot(x='name', y='gmean', hue='norm', errorbar='sd', data=res)
plt.xticks(rotation=90)
plt.title('Comparison plot before and after standardization')
plt.savefig('wykres_porownawczy_normalizacji_zadanie_5.png')
```

```python
from sklearn.metrics import make_scorer
from sklearn.pipeline import Pipeline
from tqdm.notebook import tqdm
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, cross_validate
```

```python
grid = dict(
    vote__voting=['hard', 'soft'],
    vote__tree__criterion=['gini', 'entropy'],
    vote__tree__max_depth=[2, 3, 6, 12, None],
    vote__tree__class_weight=[None, 'balanced'],
    vote__gnb__var_smoothing=[1e-9, 1e-8, 1e-7],
    vote__svc__C=[0.1, 1, 10, 100],
    vote__svc__gamma=[1, 0.1, 0.01],
    vote__svc__kernel=['rbf', 'poly', 'sigmoid']
)
k_fold = 10
args = dict(cv=k_fold, n_jobs=4, scoring=dict(gmean=make_scorer(geometric_mean_score)))
model_balanced = Pipeline([
    ('scale', StandardScaler()),
    ('vote', VotingClassifier([('tree', DecisionTreeClassifier(class_weight='balanced')), ('gnb', GaussianNB()), ('svc', SVC(class_weight='balanced'))]))
])

model_imbalanced = Pipeline([
    ('scale', StandardScaler()),
    ('vote', VotingClassifier([('tree', DecisionTreeClassifier(class_weight=None)), ('gnb', GaussianNB()), ('svc', SVC(class_weight=None))]))
])
```

```python
grid_search = RandomizedSearchCV(model_imbalanced, grid, n_iter=20, refit='gmean', **args)
tuned_imbalanced = grid_search.fit(X, y)
```

```python
grid_search = RandomizedSearchCV(model_balanced, grid, n_iter=20, refit='gmean', **args)
tuned_balanced = grid_search.fit(X, y)
```

```python
gmean_tuned_imbalanced = tuned_imbalanced.cv_results_['mean_test_gmean']
gmean_tuned_imbalanced[np.isnan(gmean_tuned_imbalanced)] = 0
print(gmean_tuned_imbalanced.max())
```

```python
gmean_tuned_balanced = tuned_balanced.cv_results_['mean_test_gmean']
gmean_tuned_balanced[np.isnan(gmean_tuned_balanced)] = 0
print(gmean_tuned_balanced.max())
```

```python
base_imbalanced = cross_validate(model_imbalanced, X, y, **args)
```

```python
print(base_imbalanced['test_gmean'].mean())
```

```python
def remove_items(test_list, item):
 
    # using list comprehension to perform the task
    res = [i for i in test_list if i != item]
 
    return res

gmean_tuned_imbalanced = remove_items(gmean_tuned_imbalanced, 0)
gmean_tuned_balanced = remove_items(gmean_tuned_balanced, 0)
```

```python
gmean_tuned_balanced_std = tuned_balanced.cv_results_['std_test_gmean']
gmean_tuned_balanced_std[gmean_tuned_balanced_std == 0] = 10
print(min(gmean_tuned_balanced_std))

gmean_tuned_imbalanced_std = tuned_imbalanced.cv_results_['std_test_gmean']
gmean_tuned_imbalanced_std[gmean_tuned_imbalanced_std == 0] = 0.9
print(min(gmean_tuned_imbalanced_std))
```

```python
data = pd.DataFrame([
    dict(name='base', gmean=base_imbalanced['test_gmean'].mean(), gmean_std=base_imbalanced['test_gmean'].std()),
    dict(name='tuned', gmean=max(gmean_tuned_imbalanced), gmean_std=min(gmean_tuned_imbalanced_std)),
    dict(name='tuned + balanced', gmean=max(gmean_tuned_balanced), gmean_std=min(gmean_tuned_balanced_std))
])
```

```python
plt.figure(figsize=(20, 12))
sns.barplot(x='name', y='gmean', data=data)
plt.ylim(0.8)
plt.errorbar(x=[0, 1, 2], y=data.gmean, yerr=data.gmean_std, fmt='none', c='k', lw=2)
plt.title('Comparison of tuning best ensemble model')
plt.savefig('wykres_porownawczy_strojenia_zadanie_6.png')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
model = tuned_balanced.best_estimator_
model.fit(X_train, y_train)
y_pr = model.predict(X_test)
cm = confusion_matrix(y_test, y_pr, labels=tuned_balanced.best_estimator_.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=tuned_balanced.best_estimator_.classes_)
plt.figure(figsize=(20, 12))
disp.plot()
plt.title('Confusion matrix of the best model from GridSearchCV')
plt.savefig('macierz_pomylek_po_tuningu_zadanie7.png')
```

```python
model = StackingClassifier([('tree', DecisionTreeClassifier()), ('svc', SVC()), ('gnb', GaussianNB())], n_jobs=4)
model.fit(X_train, y_train)
y_pr = model.predict(X_test)
cm = confusion_matrix(y_test, y_pr, labels=model.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
plt.figure(figsize=(20, 12))
disp.plot()
plt.title('Confusion matrix of the best model with custom parameters')
plt.savefig('macierz_pomylek_przed_tuningiem_zadanie7.png')
```
