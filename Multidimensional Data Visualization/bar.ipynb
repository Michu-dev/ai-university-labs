```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as mcolors
```

```python

u = np.round(np.arange(0, 1.01, 0.000005), 7)
v = np.round(np.array([np.random.uniform(0, 1 - uu) for uu in u]), 7)
w = np.round(1 - u - v, 7)
u[u < 0] = 0
v[v < 0] = 0
w[w < 0] = 0
u[u > 1] = 1
v[v > 1] = 1
w[w > 1] = 1
# func2 = v - w
# func3 = w - u

a, b, c = (0, 0), (1, 0), (1/2, np.sqrt(3)/2)
x, y = [a[0], b[0], c[0], a[0]], [a[1], b[1], c[1], a[1]]
```

```python
# Equations
func = v**(1/2) + u**(1/2) - 1
xs, ys = zip(*[(a[0] * u[i] + b[0] * v[i] + c[0] * w[i], a[1] * u[i] + b[1] * v[i] + c[1] * w[i]) for i, p in enumerate(func) if np.round(p, 2) == 0])
```

```python
func = u**2 - v
func2 = v**2 - w
func3 = w**2 - u
# Inequalities - keep an eye on sign!
# xs, ys = zip(*[(a[0] * u[i] + b[0] * v[i] + c[0] * w[i], a[1] * u[i] + b[1] * v[i] + c[1] * w[i]) for i, p in enumerate(func) if np.round(p, 2) >= 0])
# 2 functions
# xs, ys = zip(*[(a[0] * u[i] + b[0] * v[i] + c[0] * w[i], a[1] * u[i] + b[1] * v[i] + c[1] * w[i]) for i, _ in enumerate(func)
#                 if np.round(func[i], 2) <= 0 and np.round(func2[i], 2) <= 0])
# 3 functions
xs, ys = zip(*[(a[0] * u[i] + b[0] * v[i] + c[0] * w[i], a[1] * u[i] + b[1] * v[i] + c[1] * w[i]) for i, _ in enumerate(func) 
               if np.round(func[i], 2) <= 0 and np.round(func2[i], 2) <= 0 and np.round(func3[i], 2) <= 0])
```

```python
plt.plot(x, y)
plt.plot(xs, ys, 'ro', ms=4)
plt.show()
```
